```julia
using Maxima

mx= Maxima.Session()
# Definimos la puerta Hadamard
H = mx.eval("1/sqrt(2) * matrix([1, 1], [1, -1])")

# Estado inicial |0⟩
ket0 = mx.eval("matrix([1], [0])")

# Aplicamos la puerta Hadamard
result = mx.eval("H . ket0")
println("Resultado simbólico de H|0⟩:")
println(result)
```
