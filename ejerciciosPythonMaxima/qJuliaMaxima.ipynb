# Cicuitos básicos ejecutados en Maxima desde Julia

**¿Para qué?**
 
En máxima podemos aprender como es la base de la computación cuántica. Podemos multiplicar vectores por matrices que es el equivalente a poner una puerta cuántica en un qubit. Esta es la base. 
Inmediatamente después en la siguiente celda de jupyter notebook podemos ver su equivalencia en un lenguaje de alto nivel para comprobar de primera mano que el algebra lineal cuántica es la base de la computación cuántica y como se puede traducir en un lenguaje de alto nivel y ejecutar en un simulador o en un ordenador cuántico real. 


**La idea** es utilizar un lenguaje de alto nivel como Julia que además tiene aplicaciones estadísticas (que pueden utilizarse para la modelización de errores de qubits), de manera que podamos 
ejecutar comandos de maxima desde julia y luego realizar ejecuciones de programas cuánticos desde un entorno julia jupyter. 

Esto permite ver como la computación cuántica consiste en algebra lineal aplicando comandos maxima y a continuación en la celda siguiente ejecutar en lenguaje de alto nivel el circuito 
cuántico anterior que se había calculado con comandos maxima.
