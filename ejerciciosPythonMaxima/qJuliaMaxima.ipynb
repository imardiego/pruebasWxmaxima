## Julia enviando comandos a Maxima

Julia puede comunicarse con Maxima usando el paquete Maxima.jl. 

Aunque Maxima no tiene soporte nativo para computación cuántica, puedes usarlo para representar simbólicamente matrices y operaciones.


```julia
import Pkg; 
Pkg.add("Maxima")
```

```julia
using Maxima

# Definimos la puerta Hadamard
H = Maxima.eval("1/sqrt(2) * matrix([1, 1], [1, -1])")
```

```julia
using Maxima

# Definimos la puerta Hadamard
H = Maxima.eval("1/sqrt(2) * matrix([1, 1], [1, -1])")

# Estado inicial |0⟩
ket0 = Maxima.eval("matrix([1], [0])")

# Aplicamos la puerta Hadamard
result = Maxima.eval("H . ket0")
println("Resultado simbólico de H|0⟩:")
println(result)
```
